#### Load datasets from CSV

```python
users = pd.read_csv("https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv" )
```

```python
sessions = pd.read_csv("https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv")
```

```python
products = pd.read_csv("https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv" )
```

```python
transactions = pd.read_csv("https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv")
```

```python
users.head() 
```

```python
sessions.head() 
```

```python
transactions.head()
```

#### Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

```python
pd.merge(transactions,  users,  how='left', on='UserID')
```

#### Which transactions have a UserID not in users?

```python
merged_table = pd.merge(transactions,  users,  how='left', on='UserID' )
```

```python
transactions[merged_table["User"].isnull()]        # There are 10 rows in Transaction table. That is the key
```

#### Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)

```python
pd.merge(transactions,  users,  how='inner', on='UserID' )
```

#### Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

```python
pd.merge(transactions,  users,  how='outer' )
```

#### Determine which sessions occurred on the same day each user registered

```python
pd.merge(users, sessions, left_on=["UserID", "Registered"], right_on=["UserID", "SessionDate"], how="inner")
```

#### Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
# Create a dummy column and join on that
users.assign(foo="1").merge(products.assign(foo="1")).drop("foo", axis=1)[["UserID","ProductID"]]
```

#### Determine how much quantity of each product was purchased by each user

```python
products
```

```python
user_product = users.assign(foo="1").merge(products.assign(foo="1")).drop("foo", axis=1)[["UserID","ProductID"]]
```

```python
pd.merge(user_product, transactions, how="left").fillna(value=0)[["UserID","ProductID","Quantity"]]
```

#### For each user, get each possible pair of pair transactions

```python
pd.merge(transactions, transactions, on='UserID')
```

#### Join each user to his/her first occuring transaction in the transactions table

```python
transactions.groupby('UserID').first().reset_index()
```

```python
data= pd.merge(users,transactions.groupby('UserID').first().reset_index(), how='left', on='UserID' ); data
```

#### Test to see if we can drop columns

```python
my_columns = list(data.columns); my_columns
```

```python
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns)  # Thresh --> Number of non NaN values
```

```python
data.shape[0] * .9
```

```python
missing_info = list(data.columns[data.isnull().any()]); missing_info
```

```python
data.isnull().any()
```

```python
# Find the number of missing rows in a column
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0] 
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
# Calculate the percentate
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / len(data)
    print('percent missing for column {}: {}'.format( col,percent_missing))
```
